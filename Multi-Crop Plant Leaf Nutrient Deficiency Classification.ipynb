```python
# ✅ Nutrient Deficiency Prediction from Image using Ensemble (EfficientNet + ResNet + DenseNet + GradCAM)

import torch
import torch.nn.functional as F
from torchvision import models, transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os

# ✅ Load class names
class_names = ['boron', 'calcium', 'healthy', 'iron', 'magnesium', 'manganese', 'nitrogen', 'phosphorus', 'potassium', 'sulphur', 'zinc']

# ✅ Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ✅ Preprocessing
inference_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# ✅ Load each model
def load_model(model_name, model_path):
    if model_name == "efficientnet_b0":
        model = models.efficientnet_b0(weights=None)
        model.classifier[1] = torch.nn.Linear(model.classifier[1].in_features, len(class_names))
    elif model_name == "resnet50":
        model = models.resnet50(weights=None)
        model.fc = torch.nn.Linear(model.fc.in_features, len(class_names))
    elif model_name == "densenet121":
        model = models.densenet121(weights=None)
        model.classifier = torch.nn.Linear(model.classifier.in_features, len(class_names))
    else:
        raise ValueError("Unsupported model")

    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()
    model.to(device)
    return model

# ✅ Predict using ensemble
def predict_image_ensemble(image_path, models_dict):
    img = Image.open(image_path).convert('RGB')
    img_tensor = inference_transform(img).unsqueeze(0).to(device)

    outputs = []
    with torch.no_grad():
        for model in models_dict.values():
            out = F.softmax(model(img_tensor), dim=1)
            outputs.append(out)

    avg_output = torch.mean(torch.stack(outputs), dim=0)
    pred = torch.argmax(avg_output, 1).item()

    print(f"✅ Ensemble Predicted Class: {class_names[pred]}")
    return img_tensor, pred, class_names[pred]

# ✅ GradCAM helper (uses only EfficientNet for visualization)
class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.gradients = None
        self.activations = None
        target_layer.register_forward_hook(self.save_activation)
        target_layer.register_full_backward_hook(self.save_gradient)

    def save_activation(self, module, input, output):
        self.activations = output

    def save_gradient(self, module, grad_input, grad_output):
        self.gradients = grad_output[0]

    def __call__(self, input_tensor, class_idx):
        output = self.model(input_tensor)
        self.model.zero_grad()
        loss = output[0, class_idx]
        loss.backward()

        gradients = self.gradients.detach().cpu().numpy()[0]
        activations = self.activations.detach().cpu().numpy()[0]


        weights = np.mean(gradients, axis=(1, 2))
        cam = np.zeros(activations.shape[1:], dtype=np.float32)

        for i, w in enumerate(weights):
            cam += w * activations[i]

        cam = np.maximum(cam, 0)
        cam = cv2.resize(cam, (224, 224))
        cam -= cam.min()
        cam /= cam.max()
        return cam

# ✅ Visualize GradCAM
def show_gradcam(img_tensor, cam, image_path, pred_class):
    img = Image.open(image_path).resize((224, 224))
    img_np = np.array(img)

    heatmap = cv2.applyColorMap(np.uint8(255 * cam), cv2.COLORMAP_JET)
    heatmap = np.float32(heatmap) / 255
    overlay = heatmap + np.float32(img_np) / 255
    overlay = overlay / np.max(overlay)

    plt.figure(figsize=(6, 6))
    plt.imshow(overlay)
    plt.title(f"Predicted: {pred_class}")
    plt.axis('off')
    plt.show()

# ✅ USAGE
image_path = "/content/drive/MyDrive/Multi-Crop Plant Leaf Nutrient Deficiency Classification/data/val/calcium/ca115.jpg"  # replace with your test image
model_paths = {
    "efficientnet_b0": "/content/drive/MyDrive/Multi-Crop Plant Leaf Nutrient Deficiency Classification/models/best_model_efficientnet_b0.pth",
    "resnet50": "/content/drive/MyDrive/Multi-Crop Plant Leaf Nutrient Deficiency Classification/models/best_model_resnet50.pth",
    "densenet121": "/content/drive/MyDrive/Multi-Crop Plant Leaf Nutrient Deficiency Classification/models/best_model_densenet121.pth",
}

# Load all models
models_dict = {name: load_model(name, path) for name, path in model_paths.items()}

# Predict using ensemble
input_tensor, class_idx, class_label = predict_image_ensemble(image_path, models_dict)

# Show GradCAM using EfficientNet
gradcam = GradCAM(models_dict['efficientnet_b0'], models_dict['efficientnet_b0'].features[-1])
cam = gradcam(input_tensor, class_idx)
show_gradcam(input_tensor, cam, image_path, class_label)
```
